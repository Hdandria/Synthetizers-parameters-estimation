# Dataset visualizer
Print some informations and listen to some audio from the dataset

```python
import random

import h5py
import IPython.display as ipd
import matplotlib.pyplot as plt

# Change this to the path of the dataset file
file_path = "../datasets/vst_samples_example/dataset_0000.h5"
```

```python
# Load the dataset
dataset = h5py.File(file_path, "r")

print(f"Dataset loaded, contains {len(dataset['audio'])} samples")
```

## Basic informations

```python
"""Print basic info about the dataset"""
print("=" * 50)
print("DATASET SUMMARY")
print("=" * 50)

num_samples = len(dataset["audio"])
print(f"Number of samples: {num_samples}")

if num_samples > 0:
    print("KEYS:")
    for key in dataset.keys():
        print(f"{key}: {dataset[key].shape}")
```

## Listen to some audio

```python
sample_idx = random.randint(0, len(dataset["audio"]) - 1)
audio_data = dataset["audio"][sample_idx]

# Display the audio
ipd.Audio(audio_data, rate=44100)
```

```python
## View the mel spectrogram
mel_spec = dataset["mel_spec"][sample_idx][0] # Left channel
plt.imshow(mel_spec)
plt.show()
```
