```python
num_std_0 = 0
num_count_3303 = 0
for key, value in dico.items():
    if value['std'] == 0:
        num_std_0 += 1
    if value['count'] == 3303:
        num_count_3303 += 1
print(f"Number of parameters with std = 0: {num_std_0}")
print(f"Number of parameters with count = 3303: {num_count_3303}")
```

```python
# Normalize mean and std by their max values
for key, value in dico.items():
    if value['max'] == 0:
        continue
    value['mean'] /= value['max']
    value['std'] /= value['max']
print(dico['beats_per_minute']) # count, mean, std, min, max
```

```python
# Plot all the unique_count histogram
import matplotlib.pyplot as plt
unique_counts = [value['unique_count'] for value in dico.values()]
unique_counts = sorted(unique_counts)
plt.hist(unique_counts, bins=50)
plt.xlabel('Unique Counts')
plt.ylabel('Number of Parameters')
plt.title('Histogram of Unique Counts for Vital Parameters')
plt.yscale('log')
plt.show()
```
